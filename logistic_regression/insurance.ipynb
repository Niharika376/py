```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/insurance-vs-age/insurance_data.csv")
df.head(10)
```

```python
plt.scatter(df["age"], df["bought_insurance"], marker = ".", color = "green")
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[["age"]], df.bought_insurance, train_size=0.9, random_state=43)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
insurance_model = LogisticRegression()
```

```python
insurance_model.fit(x_train, y_train)
```

```python
insurance_model.predict(x_test)
```

```python
insurance_model.score(x_test,y_test)
```
